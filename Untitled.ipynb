```python
from scipy.integrate import newton_cotes
def f(x):
    return np.exp(x)
a = 0
b = np.pi
exact = np.exp(np.pi)-1
for N in [2, 4, 6, 8, 10]:
    x = np.linspace(a, b, N + 1)
    an, _ = newton_cotes(N, 1)
    dx = (b - a) / N
    quad = dx * np.sum(an * f(x))
    error = abs(quad - exact)
    print('{:2d}  {:10.9f}  {:.5e}'.format(N, quad, error))
```

```python
k, l = 2, 20
an, _ = newton_cotes(k, 1)

N = l * k
x = np.linspace(a, b, N + 1)
dx = (b - a) / N
quad = 0
for i in range(l):
    quad += dx * np.sum(an * f(x[i*k:(i+1)*k+1]))
    
error = abs(quad - exact)

print('{:2d}  {:10.9f}  {:.5e}'.format(N, quad, error))
```

```python
def boole(y, x):
    
    assert (len(x)-1)%4==0, 'len(x)-1 is not divisible by 4'
        
    return (2*(x[1]-x[0])/45) * (7*y[0] + 32*np.sum(y[1:-1:2]) + 12*np.sum(y[2:-1:4]) + 14*np.sum(y[4:-1:4]) + 7*y[-1])

boole(f(x), x) - exact
```

# Hilbert transform
## comparison between two implementations(integration by definition and Fourier transform)

```python
Ns = [12, 20, 40, 80, 160, 320, 640, 1200, 2400, 4800, 9600]
err1, err2 = [], []
t = 0.569024

for N in Ns:
    
    tau = np.linspace(-1, 1, num=N+1)
    
    u = 1 - tau**2 + 0.05*np.sin(400*tau)

    err1.append(np.trapz(u/(t-tau), tau) / np.pi)
    
    err2.append(boole(u/(t-tau), tau) / np.pi)

    
fig, ax = plt.subplots(1,1, figsize=(10,6)) 

plt.loglog(Ns, np.abs(np.array(err1)-err2[-1]))

plt.loglog(Ns, np.abs(np.array(err2)-err2[-1]))

plt.show()
```
